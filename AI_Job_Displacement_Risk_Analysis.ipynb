# Impact of AI on Diffrent Job Roles and Industry

### 1. Importing Libraries & Loading Cleaned Datasets
This block handles setting up the development environment and **loading** the dataset to execute initial structural validation checks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import random

# Loading the raw dataset into the active workspace memory dataframe
df = pd.read_csv('ai_impact_jobs_layoff_risk.csv')

# Displaying a structural slice of the data schema
print("--- Dataset Snapshot ---")
print(df.head())

print("\n--- Data Information & Schema Evaluation ---")
print(df.info())
```

### 2. Analysing Layoff Risks Distribution and Automation Impact 

##### Statistical Evaluation of Automation Thresholds vs. Employee Displacement Risk

```python
# Formatting the baseline visualization aesthetics
sns.set_theme(style="whitegrid")

# Plot 1: Loading structural distributions across different operational fields
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Industry', hue='Layoff_Risk', palette='Set2')
plt.title('Layoff Risk Distribution by Industry', fontsize=14, fontweight='bold')
plt.xlabel('Industry', fontsize=12)
plt.ylabel('Employee Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('industry_vs_layoff_risk.png', dpi=300)
plt.show()

# Plot 2: Loading operational metrics evaluating routine task vulnerabilities
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Layoff_Risk', y='Routine_Task_Percentage', hue='Layoff_Risk', palette='coolwarm', legend=False)
plt.title('Impact of Routine Tasks on Layoff Risk', fontsize=14, fontweight='bold')
plt.xlabel('Layoff Risk Level', fontsize=12)
plt.ylabel('Routine Task Percentage (%)', fontsize=12)
plt.tight_layout()
plt.savefig('routine_tasks_impact.png', dpi=300)
plt.show()
```

### REPORT of Layoff Risks Distribution and Automation Impact

<details>
<summary><b>🔍 Click Me to Expand: Data Analysis & Interpretation</b></summary>

#### 1. Layoff Risk Distribution by Industry
Looking at the count plot analyzing employee layout risk tiers across distinct sectors, several key trends emerge:
* **High-Risk Sectors:** The **Manufacturing** and **Retail** industries exhibit the highest volume of employees categorized under high layoff risks. 
* **Stable Sectors:** **Healthcare** and **Education** show significantly lower vulnerability, with the vast majority of their workforce falling into the low-risk tier.
* **Balanced Distribution:** Sectors like **Finance**, **IT**, and **Telecom** show a relatively uniform spread across high, medium, and low risk, indicating that risk in these areas might be heavily dependent on specific job roles rather than the industry as a whole.

#### 2. Evaluating Routine Task Percentage Thresholds
The box-and-whisker plot maps individual positions along the operational metrics of routine workflows against overall displacement risk:
* **High Layoff Risk:** Positions facing high displacement risks are heavily concentrated where tasks are highly repetitive, with a median routine task percentage sitting well above **70%**.
* **Low Layoff Risk:** Conversely, roles with a low layoff risk score feature a much lower reliance on repetitive workflows, maintaining a median routine task threshold closer to **30% to 40%**.
* **Statistical Variations:** The presence of outliers in the medium-risk category suggests that while routine task density is a primary driver of displacement, secondary factors (such as human interaction levels or specialized tool usage) still play a defensive role for certain positions.

---
**Conclusion:** The data strongly confirms a direct correlation between high workflow automation potential (Routine Task %) and an elevated risk of organizational layoffs, particularly within asset-heavy and operational industries like Manufacturing and Retail.

</details>

### 3. Multi-Variable Skill Attenuation Analysis

##### Analyzing the Combined Impact of Creativity and Human Interaction on Displacement Vulnerabilities

```python
# Plot 3: Loading the relationship between Creativity, Human Interaction, and Layoff Risk
plt.figure(figsize=(10, 8))

# Custom hex color 
custom_risk_colors = ["#eb0707", "#13e833", "#140cf5"]

# Loading a scatter plot to map the interaction of human and creative skills
sns.scatterplot(
    data=df, 
    x='Creativity_Requirement', 
    y='Human_Interaction_Level', 
    hue='Layoff_Risk', 
    palette=custom_risk_colors,
    alpha=0.8,
    s=100  # Sets the size of the data points to make them clearly visible
)

plt.title('The Shield Effect: How Creativity and Human Interaction Deflect Layoff Risk', fontsize=14, fontweight='bold')
plt.xlabel('Creativity Requirement Index', fontsize=12)
plt.ylabel('Human Interaction Level', fontsize=12)
plt.legend(title='Layoff Risk Level', loc='upper right')
plt.tight_layout()
plt.savefig('creativity_vs_interaction_risk.png', dpi=300)
plt.show()
```

### REPORT of Multi-Variable Skill Attenuation

<details>
<summary><b>🔍 Click Me to Expand: Creativity & Human Interaction Analysis</b></summary>

#### 3. The Shield Effect: Creativity and Human Interaction vs. Layoff Risk
This scatter plot maps individual job positions across two critical human performance metrics: **Creativity Requirement Index** (x-axis) and **Human Interaction Level** (y-axis), color-coded by **Layoff Risk Level**. The distribution reveals a clear defensive boundary against displacement:

* **The High-Risk Zone (Red Cluster):** Positions with low creativity requirements (under **40-50** on the index) are overwhelmingly classified as **High Layoff Risk**, regardless of their human interaction levels. This indicates that human interaction alone isn't enough to protect a job if the underlying tasks lack creativity.
* 
* **The Transition Zone (Purple Cluster):** A dense diagonal band of **Medium Layoff Risk** forms between the indices of **40 and 70**. In this zone, a balanced blend of rising creativity requirements and moderate human interaction acts as a partial buffer against immediate displacement.
* 
* **The Safe Haven (Green Cluster):** Once a role’s creativity requirement crosses the **65-70 threshold**, the risk drops drastically to **Low Layoff Risk**. Interestingly, high creativity provides immense protection even when human interaction scores are low.

---
#### **Conclusion:**
**Creativity** acts as the primary "shield" against layoff risks. While human interaction provides a secondary layer of defense, cognitive uniqueness and non-routine creative requirements remain the strongest deterrents to automation-driven job displacement.

</details>

### 4. Interaction Between Firm Scale, Tool Adjacency, and Workflow Automation

##### This analysis evaluates the structural relationship between operational automation, the density of specialized application integration, and organizational scale. By plotting the **Number of AI Tools Used** ($x$-axis) against the **Routine_Task_Percentage** ($y$-axis) and stratifying by **Company_Size**, we uncover the exact mechanics of corporate adaptation patterns.

```python
# Plot 4: Loading the relationship between Company Size, AI Tools, and Routine Tasks
plt.figure(figsize=(10, 6))

# Custom professional hex color palette 
company_scale_colors = ["#264653", "#f20707", "#07f21b"]

# Loading a bar plot that calculates the average routine task percentage automatically
sns.barplot(
    data=df, 
    x='Number_of_AI_Tools_Used', 
    y='Routine_Task_Percentage', 
    hue='Company_Size', 
    palette=company_scale_colors,
    errorbar=None  # Removes error bars to keep the visualization clean and simple
        )

plt.title('Average Routine Task Percentage by Number of AI Tools and Company Size', fontsize=14, fontweight='bold')
plt.xlabel('Number of AI Tools Used', fontsize=12)
plt.ylabel('Average Routine Task Percentage (%)', fontsize=12)
plt.legend(title='Company Size', loc='upper right')
plt.tight_layout()
plt.savefig('simplified_company_trends.png', dpi=300)
plt.show()
```

### REPORT of Interaction Between Firm Scale, Tool Adjacency, and Workflow Automation

<details>
<summary><b>🔍 Click to Expand: AI Tool Adoption & Firm Size Analysis</b></summary>

#### 4. Evaluating Tool Scaling and Routine Core Workflows
This grouped bar chart evaluates how the average density of routine tasks scales alongside the **Number of AI Tools Used** ($x$-axis), stratified across **Company Size** (Medium, Large, and Small):

* **The Low Adoption Threshold (0–1 Tools):** At zero or single-tool environments, the average routine task workload remains low, hovering around **30% to 35%**. This reflects operational states that haven't deeply integrated automated infrastructure yet.
 
* **The Integration Inflection Point (2–5 Tools):** A clear step-up occurs once teams cross into utilizing 2 to 5 tools. Routine task metrics jump straight to the **45% to 70%** bracket. This indicates that organizations are actively applying tooling suites to highly repetitive pipelines.
  
* **The Hyper-Automation Ceiling (6+ Tools):** When tool adoption hits 6 or more concurrent platforms, routine task percentages plateau at a massive **85%**. This represents environments where repetitive operations are completely saturated by specialized automation suites.
  
* **Company Size Uniformity:** Interestingly, across almost every single tool tier, **Medium, Large, and Small firms exhibit near-identical metrics**. The structural relationship between tool counts and workflow routinization is driven entirely by technical operational design rather than the physical size or scale of the company.

---
**Conclusion:** The step-like growth trends confirm that higher counts of integrated tools map directly onto highly structured, predictable corporate workflows. Furthermore, automation scaling dynamics remain tightly uniform across all firm sizes, showing no divergence between small startups and enterprise-level corporations.

</details>

### 5. Technological Exposure Dynamics by Job Designation

##### This analysis evaluates the structural relationship between specific professional roles and their daily operational reliance on automated systems. By plotting the **Job_Role** against the **AI_Usage_Hours_Per_Week** and arranging them hierarchically, we reveal the exact footprint of modern digital tool dependency.

```python
# Plot 5: Loading Average AI Usage Hours across different Job Roles
plt.figure(figsize=(14, 7))

# Generating a unified color gradient layout
role_colors = sns.color_palette("mako", n_colors=len(df['Job_Role'].unique()))

# Loading a bar chart sorted by average hours to show a clear hierarchy
order = df.groupby('Job_Role')['AI_Usage_Hours_Per_Week'].mean().sort_values(ascending=False).index

sns.barplot(
    data=df, 
    x='Job_Role', 
    y='AI_Usage_Hours_Per_Week', 
    hue='Job_Role',
    palette=role_colors,
    order=order,
    legend=False,
    errorbar=None
)

plt.title('Technological Exposure: Average Weekly AI Usage Hours by Job Role', fontsize=14, fontweight='bold')
plt.xlabel('Job Designation', fontsize=12)
plt.ylabel('Average AI Usage Hours Per Week', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('job_role_vs_ai_hours.png', dpi=300)
plt.show()
```

### REPORT of Technological Exposure Dynamics by Job Designation

<details>
<summary><b>▼ Click Me to Expand: Evaluating Workforce Tool Scaling and Weekly Integration Hours</b></summary>

This sorted bar chart evaluates how the average density of software tooling hours scales across twenty-four distinct professional roles, exposing clear variations in technological dependency:

* **The Core Automation Tier (7+ Hours):** Analytical, administrative, and core optimization roles such as **Accountant**, **Production Supervisor**, **Operator** and **Data Analyst** occupy the top-most bracket, crossing the 7-hour weekly threshold. This reflects intense integration of automated data-parsing infrastructure in highly structured domains.
* **The Mainstream Integration Baseline (6.5–7 Hours):** A remarkably stable plateau occurs across a broad middle tier—including **ML Engineers**, **Medical Assistants**, **Teachers**, **Research Assistants**, **Financial Analysts**, **Dispatchers**, **Inventory Analysts**, **Health Analysts**, **Software Engineers**, and **Network Engineers**. Here, automation acts as a steady workflow supplement, showing a nearly identical operational footprint across highly diverse cognitive and analytical industries.
* **The Physical & Client-Facing Buffer (Under 6.5 Hours):** Front-line management, specialized execution, and highly active physical occupations, such as **Operations Analyst**, **Supply Chain Analyst**, **Nurse**, **Quality Engineer**, **Warehouse Manager**, **Auditor**, **Store Manager**, and **Sales Associate**, sit at the lower tail of the exposure axis. Their lower hours highlight fields that still strongly require non-digital, tactical, or on-the-floor execution.

---
**Conclusion:** The step-like distribution confirms that operational exposure is remarkably uniform and high across nearly all modern professions, with every single tracked role averaging over 6 hours of AI usage per week. Furthermore, the highest scaling dynamics belong to data-heavy analytical and system-oversight positions, leaving zero divergence between traditional corporate functions and modern technical infrastructure.

</details>

### 6. Analysing "Lay Off Risks" Across Job Roles

##### Tells us about which job is highly automated and likely to be replaced by AI

```python
# Plot 6: Loading Layoff Risk distributions within specific Job Roles
plt.figure(figsize=(14, 7))

# Custom professional hex color palette matching corporate design standards
custom_risk_colors = ["#2a9d8f", "#f4a261", "#e76f51"]

# Loading a grouped count plot across all available job designations
sns.countplot(
    data=df, 
    x='Job_Role', 
    hue='Layoff_Risk', 
    palette=custom_risk_colors
)

plt.title('Risk Stratification Matrix: Layoff Risk Distribution Across Job Roles', fontsize=14, fontweight='bold')
plt.xlabel('Job Designation', fontsize=12)
plt.ylabel('Employee Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Layoff Risk Level', loc='upper right')
plt.tight_layout()
plt.savefig('job_role_vs_layoff_risk.png', dpi=300)
plt.show()
```

### Risk Stratification Matrix Across Job Roles REPORT

<details>
<summary><b>▼ Click Me to Expand: Evaluating Structural Workforce Stability and Vulnerability</b></summary>

This grouped count plot evaluates the structural breakdown of layoff risk tiers across various job designations, pinpointing where organizational vulnerabilities are most heavily concentrated:

* **High-Vulnerability Outliers:** Certain roles show an overwhelming concentration of **High Layoff Risk** (teal bars). **Operators** lead this category with the highest absolute count of high-risk employees (closing in on 450), closely followed by **Inventory Analysts**, **Warehouse Managers**, and **Supply Chain Analysts**. These roles feature highly automated logistics and repetitive operational tasks.
* 
* **The Low-Risk Defenses:** Conversely, several professions display an exceptionally strong defensive layer of **Low Layoff Risk** (light orange bars). **Teachers** dominate this group, registering the highest count of low-risk individuals (nearly 450), alongside solid stability observed in **Medical Assistants**, **Data Analysts**, and **Financial Analysts**.
* 
* **The Volatile Mid-Tier Distribution:** Positions such as **Quality Engineers**, **Research Assistants**, and **Nurses** exhibit a distinct multi-tiered distribution where **Medium Layoff Risk** values are highly elevated. This suggests that while these roles are partially insulated from instant disruption due to technical expertise, changing structural demands keep them in a volatile transition zone.

---
**Conclusion:** The stratification matrix clearly proves that layoff vulnerabilities are not evenly spread across the enterprise landscape. Physical logistics and repetitive data-tracking roles face intense, concentrated displacement exposure, while human-centric roles (Education/Healthcare) and core strategic analytical positions maintain the strongest baseline defense.

</details>
